```python
# function
def lexical_richness(text):
    return 100 * len(set(text)) / len(text)
```

```python
# Q1. long_words = 길이가 15자 이상인 단어 (text9)
# Q2. 길이가 7 이상이고 fdist가 7이상인 단어 반환

# Q1
long_words = [w for w in set(text9) if len(w) >= 15]
print(long_words)

# Q2
fdist3 = FreqDist(text9)
words = [w for w in set(text9) if len(w) >= 7 and fdist3[w] >= 7]
print(words)
```

```python
# America, america
# text1의 길이 그냥 했을 때와 lower로 했을 때의 개수 차이

a = len(set(w for w in text1))
b = len(set(w.lower() for w in text1))
c = len(set(w.lower() for w in text1 if w.isalpha()))

print(a, b, c)
```

```python
# if절: 1) 30보다 크면 '30보다 크다' 2) 20보다 크면 '20보다 크다' 3) 10보다 크면 '...' 4) 나머지 '작다'
sent = [11,21,31]

for number in sent:
    if number > 30:
        print('30보다 크다')
    elif number > 20:
        print('20보다 크다')
    elif number > 10:
        print('10보다 크다')
    else:
        print('작다')
```
