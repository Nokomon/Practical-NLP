### special Feature
- 숫자로 변환하는 작업
- 어떤 숫자로 바꿀 것인가?
사과
나는 사과를 좋아한다. -> [x,x,x,x]

-scalar: 크기를 갖는 숫자 예) 온도
-vector: 크기+방향성 예) 속도
-matrix : 행렬, 2차원
a =[1,2,3]
a = [1
    2
    3] -> aT
-tensor: 텐서, 3차원이상

- BAG OF WORDS: 단어 모음
: 단어의 출현수를 기반으로 벡터를 만듦
a = 나는 사과를 좋아한다. = [1,2,1]
b = 너는 사과를 싫어한다. = [1,2,1]
문서 =[1,2,1,1,1]
- 단어의 존재 유무를 알 수 있다. 
- 시퀀스 정보가 없어짐

```python
#feature 03 가져오기
file = 'feature ......'
with open(file) as f:
  data= [line.strip() for line in f.readlines()]
```

```python
# tokenize ->num
import nltk
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(' '.join(data))
tokens= word_tokenize(' '.join(data)
tokens=tokens.lower()
```

```python
from collections import Counter
bow = Counter(tokens)
```

```python
common = bow.most_common()
```

```python
#벡터로 만들기
doc_vector=[]
for word,count in bow.most_common():
  doc_vector.append(count)
```

```python
#각 문장 별로 벡터 만들기
sent_tokens =[word_tokenize(line.lower())for line in data]
token_counts = [Counter(token).most_common() for token in sent_tokens]

for sent in token_counts:
  vector =[]
  for word, count in sent:
    vector.append(count)
  print(vector)
```

```python
# tokenizeand tgen num
import nltk
nltk.download("punkt")
from nltk.tokenize import word_tokenize

tokens = word_tokenize(" ".join(data))
```

```python
tokens = [w.lower() for w in tokens]
```

```python
from collections import Counter
```

```python
bow = Counter(tokens)
```

```python
bow.most_common()
```

```python
vector = [j for (i, j) in bow.most_common()]
print(vector)
```
