# 의미 분석(semantic analysis)  
do, doing, does, done, ... -> do 하나로 인식하게끔  
먹는, 먹다, 먹고 -> 먹다, 먹- 등 하나로 인식하게끔 (stemming, lemmatization)

1. 차원 축소
2. 텍스트 분석에서 정확도가 향상됨  

## Stemming | 어간 추출
- 어간: 굴절어미에서 변하지 않는 부분. 동사, 형용사
  - 예: 먹
- 어미: 굴절어미에서 변하는 부분
  - 예: -는, -다, ...
- 어근: 단어 전반에 쓰임. 중심 의미, 실질적 의미를 가진 부분
  - 예: 사과 vs 풋사과
- 접사: 어근에 붙어서 단어의 뜻을 제한하는 부분
  - 예: 풋

- 규칙기반 - pre-define된 rule에 맞지않은 결과가 나올 수도 있다.
- 빠르다.
- PorterStemmer, LancasterStemmer

## Lemmatization | 표제어 추출
단어의 기본형을 찾아가는 과정.  
am, are, is -> be  

- rule-based가 아니어서 정확하다.
- lemmatized corpus가 필요 -> wordnet
- rule-based가 아니어서 pos tagging 남아 있다.


### PorterStemmer

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
porter.stem('commas')
```

```python
porter.stem('goes')   # result: 'goe' -> 규칙기반이기에 불완전하다
```

```python
text = "They are students ."

[porter.stem(w) for w in text.split()]   # token으로 입력
```
